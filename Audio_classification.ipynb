```python
def load_file(file_name, sampling_rate=16000):
    return lr.core.load(file_name, sr = sampling_rate)
```

```python
def split_files(src_file, dest_loc, n_splits=(1)):
    data,sr =load_file(src_file)
    n_splits = math.floor(len(data)/
                          (16000*1))
    if not os.path.exists(dest_loc):
        os.makedirs(dest_loc)
    dest_loc = os.path.join(dest_loc, os.path.basename(src_file))
    for i,split in enumerate(np.array_split(data, n_splits), start=1):
        lr.output.write_wav(dest_loc+"_split_"+str(i), y=split, sr=16000)
        print(dest_loc+"_split_"+str(i)+"  created!")
```

```python
for file1, file2 in zip_longest(audio_files, bg_files):
    if(file1 is not None):
        split_files(file1, "processed_audio")
    if(file2 is not None):
        split_files(file2, "processed_audio")
```

```python
def squarize(vector):
    m,n = vector.shape
    if m > n:
        pad = np.zeros(shape=(m, m-n))
        return np.c_[vector,pad]
    elif n > m:
        pad = np.zeros(shape=(n-m, n))
        return np.r_[vector,pad]
    return vector
```

```python
import os, glob, math
import numpy as np
import pandas as pd
```

```python
audio_data_path= os.path.join("/Users", "preetham", "PycharmProjects", "Audio_Classification", "data" , "*.npy" )
audio_file_dir = os.path.join("/Users", "preetham", "PycharmProjects", "Audio_Classification", "processed_audio")
audio_files = os.path.join("/Users", "preetham", "PycharmProjects", "Audio_Classification", "processed_audio","*.wav_split*")
all_files = glob.glob(audio_files)
```

```python
all_files = np.random.permutation(all_files)
train = 0.8
validation = 0.165
test = 0.35
n_train_samples = math.floor(len(all_files)*train)
n_val_samples = math.floor(len(all_files)*validation)
n_test_samples = math.floor(len(all_files)*test)

train_samples = all_files[:n_train_samples]
validation_samples = all_files[n_train_samples+1 : n_train_samples+1 + n_val_samples]
test_samples = all_files[n_train_samples+1 + n_val_samples : n_train_samples+1 + n_val_samples+1+n_test_samples]

[len(train_samples), len(validation_samples), len(test_samples)]
```

### file to label mapping

```python
csv_file_path = os.path.join("/Users", "preetham", "PycharmProjects", "Audio_Classification", "train.csv" )
csv_data = pd.read_csv(csv_file_path)
file_label_map={os.path.basename(file): csv_data[csv_data.fname == os.path.basename(file)].label.values[0] for file in all_files}
dimensions = (1,1,128,128)
```

```python
def encode_label(label, labels):
    """
    create one-hot encoding of clases
    """
    try:
        idx = labels.index(label)
        one_hot = np.zeros(len(set(labels)))
        one_hot[idx] = 1
        return one_hot
    except ValueError:
        return None
```

```python
#removing dups and converting it back to a list
class_names = list(set(csv_data.label))
nb_classes = len(class_names)
total_train = len(train_samples)
total_validation = len(validation_samples)
total_test= len(test_samples)


## train and test splitting

X_train = np.zeros((total_train, dimensions[1], dimensions[2], dimensions[3]))
Y_train = np.zeros((total_train, nb_classes))
X_val = np.zeros((total_validation, dimensions[1], dimensions[2], dimensions[3]))
Y_val = np.zeros((total_validation, nb_classes))
X_test = np.zeros((total_test, dimensions[1], dimensions[2], dimensions[3]))
Y_test = np.zeros((total_test, nb_classes))
```

```python
train_count, test_count, validation_count = (0,0, 0)
for idx, _class in enumerate(class_names):
    this_Y = np.array(encode_label(_class,class_names) )
    this_Y = this_Y[np.newaxis,:]
    for _idx, mel_file in enumerate(csv_data[csv_data.label==_class].fname.values):
        mel_data_file = os.path.join("/Users", "preetham", "PycharmProjects", "Audio_Classification","data", mel_file+".npy")
        wav_file = os.path.join(audio_file_dir, mel_file);
        if(wav_file in train_samples):
            X_train[train_count,:,:] = np.load(mel_data_file)[np.newaxis,np.newaxis,:,:]
            Y_train[train_count,:] = this_Y
            train_count +=1
        elif(wav_file in test_samples):
            if(test_count == 50):
                print(wav_file)
                print(_class,this_Y)
                print(class_names)
            X_test[test_count,:,:] = np.load(mel_data_file)
            Y_test[test_count,:] = this_Y
            test_count +=1
        elif(wav_file in validation_samples):
            X_val[validation_count,:,:] = np.load(mel_data_file)
            Y_val[validation_count,:] = this_Y
            validation_count +=1
```
